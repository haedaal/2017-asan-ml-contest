### lesion stats

```python
lesions = []
num_lesions = []
for idx, pt in enumerate(benign_patients):
    stdout.write('%d\r' % idx)
    p = load_patient(pt)
    mask = p['mask']
    label = measure.label(ndimage.binary_dilation(mask, iterations=2))
    regions = measure.regionprops(label)
    num_lesions.append(len(regions))
    for region in regions:
        lesions.append({'patient': pt, 'area': region.area, 'bbox': region.bbox})
```

```python
lesions = []
num_lesions = []
for idx, pt in enumerate(benign_patients):
    stdout.write('%d\r' % idx)
    p = load_patient(pt)
    mask = p['mask']
    for slice, m in enumerate(mask):
        regions = measure.regionprops(m)
        for region in regions:
            if (region.area < 100):
                print("")
                print(pt, slice, region.bbox, region.area)
```

```python
def getPlaneSize(lesion):
    bbox = lesion['bbox']
    xw = abs(bbox[1] - bbox[4])
    yw = abs(bbox[2] - bbox[5])
    if xw > 200 or yw > 200:
        print(lesion)
    return np.array([xw, yw]) if xw > yw else np.array([yw, xw])

area = list(map(lambda l: l['area'], lesions))
print('number of lesions in a patient', np.unique(num_lesions, return_counts=True))
plt.hist(list(map(lambda l: l['area'], lesions)))
plt.show()
print('min area : %d, max area : %d' %(np.min(area), np.max(area)))
sizes = np.array(list(map(lambda l: getPlaneSize(l), lesions)))
plt.plot(sizes[:,0], sizes[:,1], 'ro')
plt.show()
```

```python
lesions = []
num_lesions = []
for idx, pt in enumerate(malign_patients):
    stdout.write('%d\r' % idx)
    p = load_patient(pt)
    mask = p['mask']
    label = measure.label(ndimage.binary_dilation(mask, iterations=2).astype(label.dtype))
    regions = measure.regionprops(label)
    num_lesions.append(len(regions))
    for region in regions:
        lesions.append({'patient': pt, 'area': region.area, 'bbox': region.bbox})
```
